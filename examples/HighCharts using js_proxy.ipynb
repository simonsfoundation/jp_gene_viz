### Drilling down

Make a bar chart with a 
[drill down event](http://jsfiddle.net/gh/get/jquery/1.7.2/highslide-software/highcharts.com/tree/master/samples/highcharts/drilldown/async/).
We had to load the drilldown module with the supporting functionality
above before we created any charts.

The tricky part here is constructing a callback that allows
Javascript to inform Python of the drilldown event.

Also, for some reason I don't understand the ipython protocol seems
to get confused if it is not allowed to do work while the callback is paused.
In particular subsequent callbacks after the first fail.

In this case we do the ajax emulation using another thread to allow
the main kernel thread to proceed.  I've also been able to avoid the problem
by calling `get_ipython.kernel.do_one_iteration()`.

```python
sc = js_proxy.ProxyWidget()

# save some typing
elt = sc.element()
win = sc.window()

# drill down data
[name, data] = "name data".split()

drilldowns = {
        'Animals': {
            name: 'Animals',
            data: [
                ['Cows', 2],
                ['Sheep', 3]
            ]
        },
        'Fruits': {
            name: 'Fruits',
            data: [
                ['Apples', 5],
                ['Oranges', 7],
                ['Bananas', 2]
            ]
        },
        'Cars': {
            name: 'Cars',
            data: [
                ['Toyota', 1],
                ['Volkswagen', 2],
                ['Opel', 5]
            ]
        }
    }
import threading

def handle_drilldown(ident, args):
    import time, pprint
    #pprint.pprint(args.keys())
    eventinfo = args["0"]
    #pprint.pprint(("info", eventinfo))
    if not eventinfo.get("seriesOption"):
        chart = elt.highcharts()
        pointname = eventinfo["point"]["name"]
        pointobjectref = chart.get(pointname)
        series = drilldowns[pointname]
        # set up a Timer to resolve the ajax emulation
        sc(chart.showLoading('Simulating calculation/retrieval delay...'))
        t = threading.Timer(2.0, resolve, [pointobjectref, series])
        t.start()
    sc.flush()

def resolve(pointobjectref, series):
    chart = elt.highcharts()
    sc(chart.hideLoading())
    sc(chart.addSeriesAsDrilldown(pointobjectref, series))
    sc.flush()
        
drilldown_callback = sc.callback(handle_drilldown, data=None, level=3)

# for debugging...
debug_function = sc.function(["event"], "debugger;")

sc(elt.width("500px").height("400px"))
sc(elt.highcharts({
        "chart": {
            "type": 'column',
             "events": {
                 "drilldown": drilldown_callback
                    #"drilldown": debug_function
               }
        },
        title: {
            "text": 'Async drilldown'
        },
        xAxis: {
            "type": 'category'
        },

        "legend": {
            "enabled": false
        },

        "plotOptions": {
            "series": {
                "borderWidth": 0,
                "dataLabels": {
                    "enabled": true
                }
            }
        },

        "series": [{
            "name": 'Things',
            "colorByPoint": true,
            "data": [{ # NOTE: NEED TO ADD ID TO LOCATE POINTS
                "id": 'Animals',
                "name": 'Animals',
                "y": 5,
                "drilldown": true
            }, {
                "id": 'Fruits',
                "name": 'Fruits',
                "y": 2,
                "drilldown": true
            }, {
                "id": 'Cars',
                "name": 'Cars',
                "y": 4,
                "drilldown": true
            }]
        }],

        "drilldown": {
            "series": []
        }
    }))
sc.flush()
display(sc)
```
